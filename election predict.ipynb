```python
#eliminating rows with NaN values
testdf = testdf[testdf['SEX'].notna()]
testdf.AGE.astype('int64')
```

```python
#merging 2009 & 2014 dataset
traindf = pd.concat([traindf1 , traindf2])
```

```python
#label encode categorical columns of train dataset
traindf['STATE'] = LabelEncoder().fit(traindf['STATE']).transform(traindf['STATE'])
traindf['CONSTITUENCY'] = LabelEncoder().fit(traindf['CONSTITUENCY']).transform(traindf['CONSTITUENCY'])
traindf['NAME'] = LabelEncoder().fit(traindf['NAME']).transform(traindf['NAME'])
traindf['PARTY'] = LabelEncoder().fit(traindf['PARTY']).transform(traindf['PARTY'])
traindf['SEX'] = LabelEncoder().fit(traindf['SEX']).transform(traindf['SEX'])
traindf['CATEGORY'] = LabelEncoder().fit(traindf['CATEGORY']).transform(traindf['CATEGORY'])
traindf.head()
```

```python
#label encode categorical columns of test dataset
testdf['STATE'] = LabelEncoder().fit(testdf['STATE']).transform(testdf['STATE'])
testdf['CONSTITUENCY'] = LabelEncoder().fit(testdf['CONSTITUENCY']).transform(testdf['CONSTITUENCY'])
testdf['NAME'] = LabelEncoder().fit(testdf['NAME']).transform(testdf['NAME'])
testdf['PARTY'] = LabelEncoder().fit(testdf['PARTY']).transform(testdf['PARTY'])
testdf['SEX'] = LabelEncoder().fit(testdf['SEX']).transform(testdf['SEX'])
testdf['CATEGORY'] = LabelEncoder().fit(testdf['CATEGORY']).transform(testdf['CATEGORY'])
testdf.head()
```

```python
# scaling values into 0-1 range
scaler = MinMaxScaler(feature_range=(0, 1))
features = [
    'STATE', 'CONSTITUENCY', 'NAME', 'PARTY', 'SEX', 'AGE', 'CATEGORY', 'TOTAL_VOTES', 'WINNER']
traindf[features] = scaler.fit_transform(traindf[features])
testdf[features] = scaler.fit_transform(testdf[features])
testdf.head()
```

```python
# separate train features and label
y_train = traindf["WINNER"]
x_train = traindf.drop(labels=["WINNER"], axis=1)

x_test = testdf.drop(labels=["WINNER"], axis=1)
y_test = testdf["WINNER"]

# train and test knn model
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
pred = knn.predict(x_test)
print("predictions :", pred)
print("Testing Accuracy is: ", knn.score(x_test, y_test)*100, "%")
```
